```python
import numpy as np
```

```python
x = np.array([2,7,4,11,13,10,25,5,17,12,22])
```

```python
y = np.array([7,97,31,241,337,199,1249,49,577,287,967])
```

```python
#from sklearn import preprocessing
```

```python
from tensorflow.keras import models
from tensorflow.keras.layers import *
```

```python
model = models.Sequential()
```

```python
model.add(Dense(128, activation='relu', input_shape = [1]))
model.add(Dense(16, activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='rmsprop',
loss='mae',
metrics=['mae'])
```

```python
model.fit(x,y, epochs=4000)
```

```python
model.save('Math_Equation.h5')
```
