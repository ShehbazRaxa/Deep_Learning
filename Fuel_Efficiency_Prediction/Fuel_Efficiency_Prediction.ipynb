```python
dummy = pd.get_dummies(ndata.origin)
```

```python
ndata = pd.concat([ndata,dummy],axis=1)
```

```python
ndata.drop('origin',axis=1,inplace=True)
```

```python
x = ndata.drop('mpg',axis=1)
```

```python
y = ndata['mpg']
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_label,test_label = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn import preprocessing
```

```python
train_data = preprocessing.normalize(train_data)
test_data = preprocessing.normalize(test_data)
```

```python
from tensorflow.keras import models
from tensorflow.keras.layers import *
```

```python
model = models.Sequential()
```

```python
model.add(Dense(128,activation='relu',input_shape=(train_data.shape[1],)))
model.add(Dense(64,activation='relu'))
model.add(Dense(1))
```

```python
model.compile(optimizer='rmsprop',loss='mae',metrics=['mae'])
```

```python
model.fit(train_data,train_label,epochs=500)
```
