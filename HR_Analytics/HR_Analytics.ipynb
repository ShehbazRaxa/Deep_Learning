```python
import pandas as pd
```

```python
dataset = pd.read_csv('HR_comma_sep.csv')
```

```python
dataset.head()
```

```python
dataset.info()
```

```python
dataset.Department.value_counts()
```

```python
depart_dummies = pd.get_dummies(dataset.Department)
```

```python
depart_dummies.head()
```

```python
salary_dummies = pd.get_dummies(dataset.salary)
```

```python
salary_dummies.head()
```

```python
dataset.drop(['salary','Department'],axis=1,inplace=True)
```

```python
dataset.head()
```

```python
df2 = pd.concat([salary_dummies,depart_dummies,dataset],axis=1)
```

```python
df2.head()
```

```python
x = df2.drop('left',axis=1)
```

```python
x.head()
```

```python
y = df2['left']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
train_data,test_data,train_label,test_label = train_test_split(x,y,test_size=0.2)
```

```python
from tensorflow.keras import models
from tensorflow.keras.layers import *
```

```python
model = models.Sequential()
```

```python
model.add(Dense(1024,activation='relu',input_shape=(train_data.shape[1],)))
model.add(Dense(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='rmsprop',metrics=['acc'])
```

```python
model.fit(train_data,train_label, epochs=40)
```

```python
model.evaluate(test_data,test_label)
```
