```python
dummy = pd.get_dummies(car_dataset)
```

```python
dummy
```

```python
y = car_dataset['Selling_Price']
```

```python
y
```

```python
x = dummy.drop('Selling_Price',axis='columns')
```

```python
len(x.columns)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_data,test_data,train_label,test_label = train_test_split(x,y,test_size=0.2)
```

```python
from sklearn import preprocessing
```

```python
train_data = preprocessing.normalize(train_data)
test_data = preprocessing.normalize(test_data)
```

```python
len(train_data)
```

```python
len(test_data)
```

```python
from tensorflow.keras import models
from tensorflow.keras.layers import *
```

```python
model = models.Sequential()
```

```python
model.add(Dense(128,activation='relu',input_shape=(train_data.shape[1],)))
model.add(Dense(64,activation='relu'))
model.add(Dense(1))
```

```python
train_data.shape[1]
```

```python
model.compile(optimizer='rmsprop',loss='mae',metrics=['mae'])
```

```python
model.fit(train_data,train_label,epochs=10)
```

```python
model.save("Car_price_prediction.h5")
```

```python
test_pred = model.predict(test_data).flatten()
```

```python
from sklearn.metrics import mean_squared_error
```

```python
errors = mean_squared_error(test_label,test_pred)
```
